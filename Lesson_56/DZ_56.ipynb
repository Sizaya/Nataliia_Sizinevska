```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as clr
```

```python
im = cv2.imread("cat.jpg")
plt.axis('off')
plt.imshow(im)
plt.show()
```

```python
# конвертуємо функцією cvtColor BGR → RGB
im_RGB = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(im_RGB)
plt.show()
```

```python
# створюємо копію зображення
R = im_RGB.copy()
# зануляємо другий (G) та третій (B) канали (R=0, G=1, B=2)
R[:,:,(1,2)] = 0 
plt.imshow(R)
plt.show()
```

```python
# створюємо копію зображення
G = im_RGB.copy()
# зануляємо перший (R) та третій (B) канали
G[:,:,(0,2)] = 0
plt.imshow(G)
plt.show()
```

```python
# створюємо копію зображення
B = im_RGB.copy()
# зануляємо перший (R) та другий (G) канали
B[:,:,(0,1)] = 0
plt.imshow(B)
plt.show()
```

```python
img_HSV = cv2.cvtColor(im, cv2.COLOR_BGR2HSV)
```

```python
H = img_HSV.copy()
# задаємо максимальне (255) значення
# S та V каналам
H[:, :, (1,2)] = 255
# конвертуємо в RGB
H_RGB =  cv2.cvtColor(H,cv2.COLOR_HSV2RGB) 
plt.show()
plt.imshow(H_RGB)
```

```python
S = img_HSV.copy()
# задаємо H максимальне значення(179)
S[:, :, 0] = 179 
# задаємо V максимальне значення(255)
S[:, :, 2] = 255 
# конвертуємо в RGB
S_RGB =  cv2.cvtColor(S,cv2.COLOR_HSV2RGB)
plt.show()
plt.imshow(S_RGB)
```

```python
V = img_HSV.copy()
# задаємо H максимальне значення(179)
V[:, :, 0] = 179
# задаємо S значення 0
V[:, :, 1] = 0
# конвертуємо в RGB
V_RGB =  cv2.cvtColor(V,cv2.COLOR_HSV2RGB) 
plt.show()
plt.imshow(V_RGB)
```

```python
im_Lab = cv2.cvtColor(im, cv2.COLOR_BGR2LAB)
```

```python
L = im_Lab.copy()
L[:, :, (1,2)] = 0
L_RGB =  cv2.cvtColor(L,cv2.COLOR_LAB2RGB)
L_GRAY = cv2.cvtColor(L_RGB,cv2.COLOR_RGB2GRAY)
# L у чорно-білій гамі кольорів
plt.imshow(L_GRAY, cmap="gray") 
plt.show()
```

```python
a = im_Lab.copy()
a[:, :, 0] = 127
a[:, :, 2] = 0
a_RGB =  cv2.cvtColor(a,cv2.COLOR_LAB2RGB)
a_GRAY = cv2.cvtColor(a_RGB,cv2.COLOR_RGB2GRAY)
```

```python
# створення гами кольорів для відображення
cmap_a = clr.LinearSegmentedColormap.from_list('custom blue', ['Red',"Gray",'Green'], N=255)
plt.imshow(a_GRAY,  cmap=cmap_a)
plt.show()
```

```python
b = im_Lab.copy()
b[:, :, 0] = 127
b[:, :, 1] = 0
b_RGB =  cv2.cvtColor(b,cv2.COLOR_LAB2RGB)
b_GRAY = cv2.cvtColor(b_RGB,cv2.COLOR_RGB2GRAY)
```

```python
# створення гами кольорів для відображення
cmap_b = clr.LinearSegmentedColormap.from_list('custom blue', ['Yellow',"Gray",'Blue'], N=255)
plt.imshow(b_GRAY, cmap=cmap_b)
plt.show()
```
