```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
fig = plt.figure()
```

```python
cars = pd.read_csv("cars.csv")
sns.lineplot(data=cars, x="Oil", y="Cost")
```

```python
x = cars.Oil
y = cars.Cost
plt.plot(x, y, color="red")
plt.xlabel("Oil")
plt.ylabel("Cost")
```

```python
sns.scatterplot(data=cars, x="Oil", y="Speed", color="black")
```

```python
x = cars.Speed
y = cars.Load_capacity
plt.scatter(x, y, color="green")
plt.xlabel("Speed")
plt.ylabel("Load capacity")
```

```python
speed = cars.Speed
plt.hist(speed, 5, width=20, color="pink")
```
