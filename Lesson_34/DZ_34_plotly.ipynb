```python
import plotly
import plotly.graph_objs as go
import plotly.express as px
import pandas as pd
import sweetviz as sv
import ydata_profiling as yp
from plotly.subplots import make_subplots
```

```python
city = pd.read_csv('cities.csv', index_col=0)
city.head(20)
```

```python
fig = go.Figure()
fig = make_subplots(rows=1, cols=1)
fig.add_trace(go.Scatter(x=city['population'], y=city['area']))
fig.add_trace(go.Scatter(x=city['population'], y=city['average_income']))
```

```python
city_corr = city.corr()
city_corr
```

```python
fig = go.Figure()
fig.add_trace(go.Heatmap(x=city_corr.columns, y=city_corr.index, z=city_corr))
```

```python
sv.analyze(city).show_notebook()
```
