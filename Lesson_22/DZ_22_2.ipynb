```python
from matplotlib import pyplot as plt
from matplotlib_venn import venn3


september = 30

rain = 12
wind = 8
cold = 4

rain_wind = 5
rain_cold = 3
wind_cold = 2
rain_wind_cold = 1

rain_wind_day = rain_wind - rain_wind_cold
rain_cold_day = rain_cold - rain_wind_cold
wind_cold_day = wind_cold - rain_wind_cold

rain_day = rain - rain_wind_day - rain_cold_day - rain_wind_cold
wind_day = wind - rain_wind_day - wind_cold_day - rain_wind_cold
cold_day = cold - rain_cold_day - wind_cold_day - rain_wind_cold

good_weather = september - rain_wind_cold - rain_wind_day - rain_cold_day - wind_cold_day - rain_day - wind_day - cold_day


print(f'{rain_day} days in September were rainy.')
print(f'{wind_day} days in September were windy.')
print(f'{cold_day} days in September was coldy.')
print(f'The weather was good for {good_weather} days in September.')

venn3((rain_day, wind_day, rain_wind_day, cold_day, rain_cold_day, wind_cold_day, rain_wind_cold), set_labels=("Rainy Day", "Windy Day", "Cold Day"))
plt.show()
```
