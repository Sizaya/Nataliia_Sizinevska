```python
from matplotlib import pyplot as plt
from matplotlib_venn import venn3


students = 37

maths = 15
physics = 16
chemistry = 19

maths_physics = 7
maths_chemistry = 9
physics_chemistry = 6

maths_physics_chemistry = 4

maths_physics_5 = maths_physics - maths_physics_chemistry
maths_chemistry_5 = maths_chemistry - maths_physics_chemistry
physics_chemistry_5 = physics_chemistry - maths_physics_chemistry

maths_5 = maths - maths_chemistry_5 - maths_physics_5 - maths_physics_chemistry
physics_5 = physics - maths_physics_5 - physics_chemistry_5 - maths_physics_chemistry
chemistry_5 = chemistry - maths_chemistry_5 - physics_chemistry_5 - maths_physics_chemistry

at_least_one_excelent_rating = maths_physics_chemistry + maths_physics_5 + maths_chemistry_5 + physics_chemistry_5 + maths_5 + physics_5 + chemistry_5

print(f'{maths_5} students received a grade  of 5 in math.')
print(f'{physics_5} students received a grade  of 5 in physics.')
print(f'{chemistry_5} students received a grade  of 5 in chemistry.')
print(f'{at_least_one_excelent_rating} students received at least one grade 5.')


venn3((maths_5, physics_5, maths_physics_5, chemistry_5, maths_chemistry_5, physics_chemistry_5, maths_physics_chemistry), set_labels=("Maths", "Physics", "Chemestry"))
plt.show()
```
