```python
import tensorflow as tf
import tensorflow_datasets as tfds
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Завантаження датасету CIFAR-10
dataset, info = tfds.load('cats_vs_dogs', split='train', with_info=True, as_supervised=True)
```

```python
# Вибір одного зображення з датасету
image, label = next(iter(dataset))
```

```python
# Визначення функції для аугментаційних перетворень
def augment(image):
    # Дзеркальне відображення
    flipped_image = tf.image.flip_left_right(image)

    # Зміна насиченості кольорів
    saturated_image = tf.image.random_saturation(image, lower=0.5, upper=1.5)

    # Відображення у відтінках сірого
    grayscale_image = tf.image.rgb_to_grayscale(image)

    # Поворот на 90 градусів
    rotated_image = tf.image.rot90(image)

    # Обрізка по краях
    cropped_image = tf.image.random_crop(image, size=[100, 100, 3])

    # Зміна яскравості
    brightness_image = tf.image.random_brightness(image, max_delta=0.2)

    return flipped_image, saturated_image, grayscale_image, rotated_image, cropped_image, brightness_image
```

```python
# Застосування аугментаційних перетворень до зображення
flipped_image, saturated_image, grayscale_image, rotated_image, cropped_image, brightness_image = augment(image)
```

```python
# Виведення оригінального та перетворених зображень
plt.figure(figsize=(10, 10))
```

```python
# Оригінальне зображення
plt.subplot(2, 3, 1)
plt.imshow(image)
plt.title("Original")
plt.axis('off')
```

```python
# Дзеркальне відображення
plt.subplot(2, 3, 2)
plt.imshow(flipped_image)
plt.title("Flipped")
plt.axis('off')
```

```python
# Зміна насиченості кольорів
plt.subplot(2, 3, 3)
plt.imshow(saturated_image)
plt.title("Saturated")
plt.axis('off')
```

```python
# Відображення у відтінках сірого
plt.subplot(2, 3, 4)
plt.imshow(grayscale_image, cmap='gray')
plt.title("Grayscale")
plt.axis('off')
```

```python
# Поворот на 90 градусів
plt.subplot(2, 3, 5)
plt.imshow(rotated_image)
plt.title("Rotated")
plt.axis('off')
```

```python
# Обрізка по краях
plt.subplot(2, 3, 6)
plt.imshow(cropped_image)
plt.title("Cropped")
plt.axis('off')
```
